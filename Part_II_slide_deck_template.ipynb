```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('fordgobike_cleaned.csv')
```

## GRAPH 1

> The visualisation below shows that majority of the bike users are male. Ford bike can increase revenue by targeting more adverts on avenues that has a large number of males being involved.



```python


# Checking to discover the gender with the highest number of trips
sorted_counts = df['member_gender'].value_counts()

plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False)
white_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(white_circle)
plt.title('THE TOTAL NUMBER OF TRIPS BASED ON GENDER', fontsize=16, fontweight = 'bold')
plt.axis('square')
plt.show()
```

## GRAPH 2

> The second visualisation shows that being a subcriber comes with the benefit of bike sharing. This implies that people who often take the regular rides as customer can convert to subscribe, which eventually increases Fordbike revenue.

```python
# Checking what user type had the higher number of bike sharing 
plt.figure(figsize = [10, 8])
sb.countplot(data = df, x = 'user_type', hue='bike_share_for_all_trip')
plt.title('NUMBER OF TRIPS BASED ON USER TYPE AND BIKE SHARING')
plt.ylabel('number of trips')
plt.show();
```

## GRAPH 3

> The third visualisation shows that subscribers often have a shorter trip duration. This could be as a result of having the benefit of a better route due to their subscription. This can be viewed as a benefit for individuals to convert from customers to subscribers. 

```python
plt.figure(figsize = [10, 8])
sb.barplot(data=df, x= 'user_type', y='duration_min');
plt.title('TRIP DURATION BY USER TYPE', fontsize=15, fontweight='bold')
plt.xlabel('user_type')
plt.ylabel('duration_min');
plt.show();
    
```
